```python
# Remove all rows with NA
clean_df = df.dropna()
clean_df.tail()
```

```python
# Get the max of a column
clean_df['Starting Median Salary'].max()
```

```python
# Get the index for the row withthe highest column value
clean_df['Starting Median Salary'].idxmax()
```

```python
# View the row associated with an index value
clean_df.loc[43]
```

**Major with the highest mid-career salary (after 10+ years experience)**

```python
clean_df.loc[clean_df['Mid-Career Median Salary'].idxmax()]
```

**Major with the lowest starting salary**

```python
clean_df.loc[clean_df['Starting Median Salary'].idxmin()]
```

```python
clean_df.loc[clean_df['Mid-Career Median Salary'].idxmin()]
```

**Add the difference between the 90th and 10th percentile of the mid-career columns to the original dataframe**

```python
spread_series = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Mid-Career Spread', spread_series)
clean_df
```

Sort the dataframe by the spread column

```python
spread_df = clean_df.sort_values('Mid-Career Spread')
spread_df[['Undergraduate Major', 'Mid-Career Spread']].head()
```

Use groupby to find the mean salary by major group

```python
pd.options.display.float_format = '{:,.2f}'.format  # change float format to return only the last 2 decimals
clean_df.groupby('Group').mean()
```
